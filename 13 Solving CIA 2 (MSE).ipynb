Q3. Simple Data Cleaning (James Bond dataset) Rename the column Bond Actor Salary to Bond_Salary.
1.	Convert Bond_Salary to numeric using errors='coerce'.
2.	Fill missing values in Bond_Salary with 0.

```python
bond=bond.rename(columns={"Bond Actor Salary":"Bond_Salary"})
bond
```

```python
bond["Bond_Salary"]=pd.to_numeric(bond["Bond_Salary"],errors="coerce")
```

```python
bond["Bond_Salary"]=bond["Bond_Salary"].fillna(0)
```
