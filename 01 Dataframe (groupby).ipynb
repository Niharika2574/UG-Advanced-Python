```python
#To create your own dataset without having to import it by using dataframe. Here we are creating categories like (region,country) and values like (year,sales,profit)
```

```python
data={"Region":["Asia","Asia","Europe","Europe","America","America"],"Country":["China","India","France","Italy","USA","Canada"],"Year":[2025,2023,2024,2022,2021,2025],"Sales":[20000,30000,40000,50000,60000,70000],"Profit":[300,400,500,600,700,800]}
```

```python
df=pd.DataFrame(data) #"." + tab to select DataFrame
df
```

```python
df["Country"][1]
```

```python
#groupby is a function used to split a DataFrame intro groups based on one or more criteria.
```

```python
df.groupby("Region")["Sales"].mean()
```

```python
df.groupby("Region")["Sales"].sum()
```

```python
df.groupby("Region")["Sales"].count()
```

```python
df.groupby("Region")["Sales"].min()
```

```python
df.groupby("Region")["Sales"].max()
```
