#Q6) You have the following DataFrame:
Name	Team	Salary	Profit
A	X	50000	45
B	X	55000	30
C	Y	52000	60
D	Y	53000	25
Write Python code to:
1	Group the data by Team
2	Calculate average salary and total profit for each team
3	Display the result

```python
data={"Name":["A","B","C","D"],"Team":["X","X","Y","Y"],"Salary":[50000,55000,52000,53000],"Profit":[45,30,60,25]}
data
```

```python
df=pd.DataFrame(data)
df
```

```python
#Q6)1)
```

```python
grouped=df.groupby("Team")
grouped
```

```python
#Q6)2)
```

```python
average_salary=df.groupby("Team")["Salary"].mean()
average_salary
```

```python
#OR
```

```python
average_salary=grouped["Salary"].mean()
average_salary
```

```python
average_profit=df.groupby("Team")["Profit"].sum()
average_profit
```

```python
#OR
```

```python
total_profit=grouped["Profit"].sum()
total_profit
```

```python
#Q6)3)
```

```python
result=pd.DataFrame({"Average_Salary":average_salary,"Total_Profit":total_profit})
result
```
