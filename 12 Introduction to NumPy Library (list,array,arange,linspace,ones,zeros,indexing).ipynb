```python
#selecting from numpy array (indexing)
```

```python
first_ten=np.arange(0,10)
print(first_ten)
```

```python
first_ten[3]
```

```python
first_ten[3:5] #starting from index 3 (including) to index 5 (excluding)
```

```python
first_ten[:2] #starting from the beginning to index 2 (excluding)
```

```python
first_ten[6:] #starting from index 6 (including) to the end
```

```python
first_thirty=np.linspace(0,29,30)
three_by_ten=first_thirty.reshape(3,10)
print(three_by_ten)
print(three_by_ten.shape)
```

```python
three_by_ten[1] #here only the first row will be printed
```

```python
three_by_ten[1,5] #to print only 15 from the dataset by inserting the index number of the matrix
```

```python
three_by_ten[0:2,1:3]
```
