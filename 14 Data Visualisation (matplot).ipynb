```python
def rank_performance(stock_price):
    if stock_price <= 50:
        return "Poor"
    elif stock_price > 50 and stock_price <= 100:
        return "Satisfactory"
    else:
        return "Excellent"
ibm["Close"].apply(rank_performance)
```

```python
ibm["Close"].apply(rank_performance).value_counts()
```

```python
ibm["Close"].apply(rank_performance).value_counts().plot(kind="barh") #barh gives the graph horizontally
```

```python
ibm["Close"].apply(rank_performance).value_counts().plot(kind="pie")
```
