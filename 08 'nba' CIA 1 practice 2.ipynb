```python
#drop whichever/selective rows have "na".
```

```python
nba.dropna() #dropna drops all rows and columns with "na", thus, not selectively.
```

```python
nba.dropna(how="all") #dropna+how drops only those rows or columns that have "na" everywhere.
```

```python
nba.dropna(how="any") #drops any columns/rows with na.
```

```python
nba.dropna(subset=["College"]) #College is a subset of dataset nba. This drops rows with na in College specifically.
```

```python
#Note: to save these changes, save the dataset with a new variable name.
```
