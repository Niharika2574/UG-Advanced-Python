```python
#Passing an argument and parameter as index and dataset
```

```python
days=["monday","tuesday","wednesday","thursday","friday","saturday","sunday"]
days
```

```python
fruits=["mango","kiwi","orange","melon","starfruit","strawberry","plum"]
fruits
```

```python
pd.Series(days,fruits)
```

```python
pd.Series(fruits,days)
```

```python
pd.Series(index=days,data=fruits) #to define the index and data
```
